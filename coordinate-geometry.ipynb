```python
import matplotlib.pyplot as plt

# Define points A, B, and C
point_a = (3, 4)
point_b = (-5, 7)
point_c = (2, 5)

# Calculate the slope of line AB
slope_ab = (point_b[1] - point_a[1]) / (point_b[0] - point_a[0])


# Define the equation of line AB
def line_ab(x):
    return slope_ab * (x - point_a[0]) + point_a[1]


# Define the equation of the line parallel to AB passing through C
def line_parallel_to_ab(x):
    return slope_ab * (x - point_c[0]) + point_c[1]


# Generate x values for plotting
x_values = range(-10, 10)

# Plot line AB
plt.plot(x_values, [line_ab(x) for x in x_values], label="Line AB")

# Plot points A and B
plt.plot(point_a[0], point_a[1], "bo", label="Point A")
plt.plot(point_b[0], point_b[1], "bo", label="Point B")

# Plot point C
plt.plot(point_c[0], point_c[1], "ro", label="Point C")

# Plot line parallel to AB passing through C
plt.plot(
    x_values,
    [line_parallel_to_ab(x) for x in x_values],
    label="Line parallel to AB through C",
)

# Add annotations for points
plt.text(
    point_a[0],
    point_a[1],
    f"({point_a[0]}, {point_a[1]})",
    verticalalignment="bottom",
    horizontalalignment="right",
)
plt.text(
    point_b[0],
    point_b[1],
    f"({point_b[0]}, {point_b[1]})",
    verticalalignment="bottom",
    horizontalalignment="right",
)
plt.text(
    point_c[0],
    point_c[1],
    f"({point_c[0]}, {point_c[1]})",
    verticalalignment="bottom",
    horizontalalignment="right",
)

# Add labels and legend
plt.xlabel("x")
plt.ylabel("y")
plt.title("Line AB, Points A, B, and C")

# Move legend outside the plot
plt.legend(loc="center left", bbox_to_anchor=(1, 0.5))

# Set axis limits for proper scaling
plt.xlim(-10, 10)
plt.ylim(0, 10)

# Draw x and y axes
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)

# Set transparency of grid lines
plt.grid(True, alpha=0.3)

# Hide the border box
plt.box(False)

# Show plot
plt.axis("equal")
plt.show()
```
