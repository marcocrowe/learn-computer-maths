Given the following functions:

$f(x) = 3x-5$  
$g(x) = \frac{x+3}{5}$  
$h(x) = x^2 + 4$

$g(-8) = 3(-8) - 5 = -24 - 5 = -29$

$f(h(x)) = f(x^2 + 4) = 3(x^2 + 4) - 5 = 3x^2 + 12 - 5 = 3x^2 + 7$

$f^{-1}(7)$
$f^{-1}(x)= \frac{x+5}{3}$
$f^{-1}(7) = \frac{7+5}{3} = \frac{12}{3} = 4$

the value of $x$ that makes $f(x) = g(x)$

$3x - 5 = \frac{x+3}{5} = \\ 5(3x - 5) = x + 3 = \\ 15x - 25 = x + 3 =\\ 14x = 28 = \\x = 2$

```python
def f(x):
    return x * 3 - 5


def g(x):
    return (x + 3) * 5


def h(x):
    return x ^ 2 + 3
```
