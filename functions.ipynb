```python
class Line:
    def __init__(
        self, x_values, y_values, color="blue", line_style="solid", line_width=1
    ):
        self.__x_values = x_values
        """The x values of the line."""
        self.__y_values = y_values
        """The y values of the line."""
        self.__color = color
        """The color of the line."""
        self.__line_style = line_style
        """The line style of the line."""
        self.__line_width = line_width
        """The line width of the line."""

    @property
    def x_values(self):
        """The x values of the line.

        Returns:
            _type_: _description_
        """
        return self.__x_values

    @property
    def y_values(self):
        """The y values of the line.

        Returns:
            _type_: _description_
        """
        return self.__y_values

    @property
    def color(self):
        """The color of the line.

        Returns:
            _type_: _description_
        """
        return self.__color

    @property
    def line_style(self):
        """The line style of the line.

        Returns:
            _type_: _description_
        """
        return self.__line_style

    @property
    def line_width(self):
        """The line width of the line.

        Returns:
            _type_: _description_
        """
        return self.__line_width
```

```python
def graph_function(lines: list, title, x_label, y_label):
    # Plot the sine wave for the range from -180 to 360
    for line in lines:
        plt.plot(
            line.x_values,
            line.y_values,
            color=line.color,
            linestyle=line.line_style,
            linewidth=line.line_width,
        )
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)

    # Add vertical line at x=0 and horizontal line at y=0
    plt.axhline(0, color="black", linewidth=0.5)
    plt.axvline(0, color="black", linewidth=0.5)

    # Set x ticks at every 90 degrees and move ticks to the center
    plt.xticks(arange(-180, 361, 90), rotation=0, ha="center")
    plt.yticks(arange(-1, 1.1, 0.5), rotation=90, va="center")

    # Remove the box around the plot
    plt.gca().spines["top"].set_visible(False)
    plt.gca().spines["right"].set_visible(False)

    # Adjust the position of the x-axis and y-axis lines
    plt.gca().spines["left"].set_position("zero")
    plt.gca().spines["bottom"].set_position("zero")
    plt.gca().spines["left"].set_color("gray")
    plt.gca().spines["bottom"].set_color("gray")

    plt.tight_layout()
    plt.show()
```

```python
x_values = linspace(0, 360, 1_000)  # Range from -180 to 360
cos_line = Line(x_values, 2 * sin(x_values * pi / 180), color="red")
sin_line = Line(x_values, 4 * sin(2 * x_values * pi / 180), color="blue")
# Define the x-axis values
# Generate the y-axis values (sine values)
graph_function([cos_line, sin_line], "Waves on Scale -180 to 360", "X", "Y")
```

```python
# Sin2x
x_values = linspace(-180, 360, 1_000)  # Range from -180 to 360
y_values = sin(2 * x_values * pi / 180)  # Convert degrees to radians for x2
graph_function(x_values, y_values, "Sine Wave on Scale -180 to 360", "X", "Y")
```

```python
# 2Sin2x
x_values = linspace(-180, 360, 1_000)  # Range from -180 to 360
y_values = 2 * sin(2 * x_values * pi / 180)  # Convert degrees to radians for x2
graph_function(x_values, y_values, "Sine Wave on Scale -180 to 360", "X", "Y")
```

```python
# Define the x-axis values
x_values = linspace(-180, 360, 1_000)  # Range from -180 to 360
# Generate the y-axis values (sine values)
y_values = cos(x_values * pi / 180)  # Convert degrees to radians for x2
graph_function(x_values, y_values, "Cosine Wave on Scale -180 to 360", "X", "Y")
```
